## Grupo I:

```python
#Carga del archivo
datos=loadtxt('GravI.txt')
grav=datos[:,3]
tiempo=datos[:,11]
id=datos[:,1]
altura=datos[:,2]

#Listas con archivos corregidos
GI=[]
TI=[]
IDI=[]
ALTI=[]
i=0

#Correcion de Datos replicados
while i<=len(id)-1:
    suma=0
    counter=0
    for j in range(i,len(id),1):
        if id[i]==id[j]:
            suma=suma+grav[j]
            counter=counter+1
        else:
            break #Interrumpe el ciclo cuando no encuentra mas valores repetidos
        
    IDI.append(id[i])
    GI.append(suma/counter)
    TI.append(tiempo[i])
    ALTI.append(altura[i])
    i=i+counter
    
#Calculo de la pendiente General
m=(GI[len(GI)-1]-GI[0])/(TI[len(TI)-1]-TI[0]) #Calcula la pendiente de los datos de la estacion base

#Correcion Deriva
GID=[]
for i in range(0,len(GI),1):
    GID.append(GI[i]-(m*TI[i])) #Lista con Correccion de deriva
```

```python
#Obtener datos de Linea 1  y Linea 2
GID1=[] #Gravedad linea 1
GID2=[] #Gravedad linea 2
TID1=[] #Tiempos linea 2
TID2=[] #Tiempos linea 1
IDI1=[]
IDI2=[]
ALTI1=[]
ALTI2=[]
#Se elimina la base del campo de la lista
for i in range(0,len(GID)):
    if IDI[i]<=6.1100001 or IDI[i]==6.2500000 and IDI[i]<>6.2600002:
        if IDI[i]<>6.0100002:
            GID1.append(GID[i])
            TID1.append(TI[i])
            IDI1.append(IDI[i])
            ALTI1.append(ALTI[i])
        
    elif IDI[i]>=6.1100001 or IDI[i]<>6.2500000 and IDI[i]<>6.0100002:
        if IDI[i]<>6.2600002:
            GID2.append(GID[i])
            TID2.append(TI[i])
            IDI2.append(IDI[i])
            ALTI2.append(ALTI[i])
    
#Correccion Deriva local

#Linea1
mI1=(8205.069220979718-7454.4931301061079)/(17.17500000000000-9.1444444439999995) #Calcula la pendiente de los datos de la estacion base
for i in range(0,len(GID1),1):
    GID1[i]=(GID1[i]-(mI1*TID1[i])) #Lista con Correccion de deriva

#Linea2
mI2=(7987.6393533775117-7207.6613165220242)/(16.561111109999999-14.45833333) #Calcula la pendiente de los datos de la estacion base
for i in range(0,len(GID2),1):
    GID2[i]=(GID2[i]-(mI2*TID2[i])) #Lista con Correccion de deriva
    
```

### Linea 1:

### Correccion de Deriva.

```python
figure(1) #Grafica con correcion
plot(TID1,GID1)
xlabel('Tiempo (t)')
ylabel('Gravedad (mgal)')
title('Grafica Gravedad vs Tiempo (correccion de deriva)')
```

### Correccion Free-Air.

```python
#Free-Air  para la linea 1
GI1FD=[]
for i in range(0,len(GID1),1):
    deltaH=ALTI1[i]-439.477
    correccion=0.3086*deltaH
    GI1FD.append(GID1[i]+correccion)
```

```python
figure(2) #Grafica con correcion
scatter(ALTI1,GI1FD)

xlabel('Altura (m)')
ylabel('Gravedad (mgal)')
title('                           Grafica Gravedad vs Tiempo (correccion Free-Air)')
```

```python
#Bouguer para Linea 1
GI1FDB=[]
ro=2
for i in range(0,len(GID1),1):
    deltaH=(-1)*(ALTI1[i]-439.477)
    correccion=0.04192*ro*deltaH
    GI1FDB.append(GI1FD[i]+correccion)
```

### Correccion de Bouguer:

```python
figure(2) #Grafica con correcion
scatter(ALTI1,GI1FDB)
xlabel('Altura (m)')
ylabel('Gravedad (mgal)')
title('                           Grafica Gravedad vs Tiempo (correccion Bouguer)')
```

## Grupo E:

```python
#Carga del archivo
datos=loadtxt('GravE.txt')
grav=datos[:,3]
tiempo=datos[:,11]
id=datos[:,1]

#Listas con archivos corregidos
GE=[]
TE=[]
IDE=[]
i=0

#Correcion de Datos replicados
while i<=len(id)-1:
    suma=0
    counter=0
    for j in range(i,len(id),1):
        if id[i]==id[j]:
            suma=suma+grav[j]
            counter=counter+1
        else:
            break #Interrumpe el ciclo cuando no encuentra mas valores repetidos
        
    IDE.append(id[i])
    GE.append(suma/counter)
    TE.append(tiempo[i])
    i=i+counter
    
#Calculo de la pendiente
m=(GE[len(GE)-2]-GE[1])/(TE[len(TE)-2]-TE[1]) #Calcula la pendiente de los datos de la estacion base

#Correcion Deriva
GED=[]
TED=[]
IDE1=[]
for i in range(0,len(GE),1):
    GE[i]=(GE[i]-(m*TE[i])) #Lista con Correccion de deriva
    
#Se elimina la base del campo de la lista
for i in range(0,len(GE)):
    if IDE[i]<>3.1400001 and IDE[i]<>3.0999999:
        GED.append(GE[i])
        TED.append(TE[i])
        IDE1.append(IDE[i])
```

## Linea 1:

### Correccion Deriva.

```python
figure(1) #Grafica con correcion
plot(TED,GED)
xlabel('Tiempo (t)')
ylabel('Gravedad (mgal)')
title('                       Grafica Gravedad vs Tiempo (correccion de deriva)')
```

### Correccion Free-Air.

Para esta prospeccion gravimetrica, solo podremos contar con la correccion de deriva, debido a que no se realizaron mediciones de altitud con el GPS diferencial, lo que hace imposible realizar correcciones de Free-Air.

### Correccion de Bouguer:

Para esta prospeccion gravimetrica, no es posible calcular la correccion de Bougue, debido a que no se realizaron mediciones de altitud con el GPS diferencial, asimismo, no hay localizacion del punto de la medidicion por lo tampoco tenemos informacion de densidades.
